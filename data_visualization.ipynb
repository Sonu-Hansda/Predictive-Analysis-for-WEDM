```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
# Load the synthetic dataset
df = pd.read_csv('synthetic_wedm_data.csv')

# 1. Histogram of all features
df.hist(bins=30, figsize=(14, 10), color='skyblue', edgecolor='black')
plt.suptitle('Distribution of Parameters and Responses', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
df.head()
```

```python
# 2. Pairplot to see parameter interactions
sns.pairplot(df[['Ton', 'Toff', 'IP', 'SV', 'WF', 'WT', 'CR', 'SR']], diag_kind='kde')
plt.suptitle('Pairwise Relationships', y=1.02)
plt.show()
```

```python
# 3. Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
